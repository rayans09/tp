```python
#charger les biblithèques numpy et pandas
import pandas as pd
import numpy as np
```

```python
#1.1 lire la base de données à partir du fichier csv phase1
ph1=pd.read_csv('Desktop\myL3\IA\TP_IA\Phase1.csv')
ph1.info()
```

```python
#afficher les 5 premières lignes
ph1.head()
```

```python
#Combien de valeur null par colonne
ph1.isnull().sum()
```

```python
#appliquer un encodage one hoting coding sur les colonnes de type catégorique
ph1=pd.get_dummies(ph1) 
ph1.head()
```

```python
del  ph1['class_normal'] #supprimer la colonne class_normal# du dataframe
```

```python
ph1.info()
#Afficher les statistiques de la base en utilisant la fonction describe()
ph1.describe()
```

```python
X = ph1.iloc[:,:-1] #Récupérer toutes les lignes et toutes les colonnes sauf la dernière colonne de la base
y = ph1['class_anomaly'] #Récupérer la colonne outcome qui représente le target et stocker la dans la vaiable y
```

```python
#Diviser la base de donnée en deux sous-ensembles test et train avec une taille de 30%, 70%, respectivement
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, stratify=y)
```

```python
#Charger le grid search qui cherche les paramètres optimaux d'un modèle
from sklearn.model_selection import GridSearchCV
#Charger le classificateur KNN
from sklearn.neighbors import KNeighborsClassifier
grid_params={
    'n_neighbors':[3,5,7,9,11],
    'metric':['euclidean','manhattan']
}

#Créer un objet grid search avec les paramètres et le modèle à tester
gs=GridSearchCV(
    KNeighborsClassifier(),
    grid_params,
    cv=3 #nombre de folds dans la cross validation
)
gs_results=gs.fit(X_train, y_train) #lancer le grid search et la cross validation
```

```python
gs_results.best_params_ #Afficher les meilleurs paramètres trouvés
```

```python
#qst 1.2 Créer une instance knn avec les paramètres optimaux trouvés
knn= KNeighborsClassifier(n_neighbors=3,metric='manhattan')
#Lancer l'apprentissage
knn.fit(X_train, y_train)#entraîner le modèles
y_pred=knn.predict(X_test) #evaluer le modèles
```

```python
#Afficher la matrice de confusion
from sklearn.metrics import  confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
#afficher la précision et le rappel
from sklearn.metrics import precision_score, recall_score
print('precision=',precision_score(y_test,y_pred))
print('recall=',recall_score(y_test,y_pred))
```

```python
#Appliquer une cross validation avec le knn ayant les paramètres optimaux
from sklearn.model_selection import cross_val_score
#create a new KNN model
knn_cv = KNeighborsClassifier(n_neighbors=3, metric='manhattan')#train model with cv of 5 
cv_scores = cross_val_score(knn_cv, X, y, cv=5)#print each cv score (accuracy) and average them
print(cv_scores)
print('f1={}'.format(np.mean(cv_scores)))
```
